```python
import pandas as pd
import sqlite3
from   sqlite3 import Error
```

# 2.CREATING A TABLE IN SQL WITH CSV FILE DATA

```python
# Crie uma conexão entre meu servidor e um banco de dados que eu acabei de criar
def create_connection():
    conn = None;
    try:
        conn = sqlite3.connect('database.db')
        print(f'Conexão bem sucedida com sqlite {sqlite3.version}')
    except Error as e:
        print(f'Erro{e} na conexão')
    return conn
```

```python
def execute_query(conn, query):
    try:
        df = pd.read_sql(query,conn)
        display(df)
    except Error as e:
        print(f'Erro {e} na execução das consultas')
```

```python
# chamei a conexão
conn = create_connection()
```

```python
# Crie um cursor dentro dessa conexão que vai me dar a possibilidade de fazer várias coisas. Executar consultas e buscar informações.
cur = conn.cursor()
```

```python
# Lendo os dados que eu quero colocar na minha tabela
data = pd.read_csv('data/Cleaned_Laptop_data1.csv')
```

```python
# Vou criar uma tabela
laptop = 'laptop'
```

```python
# Criando uma tabela no SQL com os dados de um arquivo CSV 
data.to_sql(laptop, conn, if_exists='replace', index=False)
```

# 3.QUERY. ANSWERING BUSSINEIS QUESTIONS WITH SQL 

```python
# Todas as linhas e colunas da tabela.
with conn:
    query0 = """  
    SELECT *
    FROM laptop; 
    """
    execute_query(conn, query0)
```

```python
# FILTRO POR LINHA
with conn:
    query1 = """   
    
    SELECT brand, model, processor_brand
    FROM laptop
    WHERE processor_brand = 'AMD';
    
    """
    execute_query(conn, query1)
```

## 3.1 QUERY

```python
# Which brand and model combination has the highest star rating
with conn:
    query2 = """  
    SELECT brand, model, star_rating
    FROM laptop
    GROUP BY brand, model
    ORDER BY star_rating DESC; 
    """
    execute_query(conn, query2)
```

```python
# Which brand has the highest number of laptop models listed in the database?
with conn:
    query3 = """  
    SELECT brand, count (model) AS model_numbers
    FROM laptop
    GROUP BY brand
    ORDER BY model_numbers DESC;
    """
    execute_query(conn, query3)
```

```python
# Which operating system (e.g., Windows, macOS, Linux) is most commonly pre-installed on the laptops?
with conn:
    query5 = """  

    SELECT os, COUNT (os) AS os_installed
    FROM laptop
    GROUP BY os
    ORDER BY os_installed DESC ;
    
    """
    execute_query(conn, query5)
```

```python
# What is the average SSD and HDD capacity among the laptops?
with conn:
    query6 = """  

    SELECT avg (ssd), avg (hdd) 
    FROM laptop
    
    """
    execute_query(conn, query6)
```

```python
# Which brand has the highest sales volume, and what is its market share?
with conn:
    query7 = """  

    SELECT brand, COUNT(*) as sales_volume
    FROM laptop
    GROUP BY brand
    ORDER BY sales_volume DESC
    LIMIT 1
    
    """
    execute_query(conn, query7)
```

```python
# What is the average price of laptops in the database?

with conn:
    query8 = """  

    SELECT avg(latest_price * 0.061) as average_price_laptops
    FROM laptop
    
    """
    execute_query(conn, query8)
```

```python
# Which laptops have the highest star ratings?

with conn:
    query9 = """  

    SELECT brand, max (star_rating) as hig_star_rating
    FROM laptop
    GROUP BY brand
    ORDER BY hig_star_rating DESC ;
    
    """
    execute_query(conn, query9)
```
